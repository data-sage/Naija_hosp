```python
hosp_table = soup.find('table', attrs={'id': table_id})
df_list = pd.read_html(str(hosp_table))
```

```python
df_list = list(range(1,82))
df_list
```

```python
for i in range(40,82):
    url='https://hfr.health.gov.ng/facilities/hospitals-search?_token=SXanE0O9O2yMDcEcVuCrHJIOrsrjwt6IsuJ6P8FG&state_id=1&lga_id=1&ward_id=0&facility_level_id=0&ownership_id=0&operational_status_id=1&registration_status_id=0&license_status_id=0&geo_codes=0&service_type=0&service_category_id=0&entries_per_page=500&page={}'.format(i)
    table_id = 'hosp'
    response = requests.get(url)
    soup = bs(response.content, 'html.parser')
    hosp_table = soup.find('table', attrs={'id': table_id})
    df_list[i-1] = pd.read_html(str(hosp_table))
```

```python
df_list
```

```python
df_final = df_list
```

```python
df_final
```

```python
df_sol = []
```

```python
for i in range(len(df_final)):
    for j in range(len(df_final[i])):
        df_sol.append(df_final[i][j])
    
```

```python
df_sol
```

```python
df = pd.concat(df_sol)
```

```python
df
```

```python
df.tail(20)
```
